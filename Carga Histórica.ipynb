```python
# Importamos las librerías con las que estaré trabajando
import requests
import json
import sqlite3
```

```python
# Creamos una conexión con base de datos creada para almacenar historia desde el 01/01/2022
con = sqlite3.connect('MisInversiones.db')
# Creamos el curso para interactuar con los datos
cursor = con.cursor()
# Creamos la tabla que almacerá la historia indicada
cursor.execute(
    '''CREATE TABLE Stocks(
        stock TEXT NOT NULL,
        volumen FLOAT,
        precio_promedio FLOAT,
        precio_apertura FLOAT,
        precio_cierre FLOAT,
        precio_maximo FLOAT,
        precio_minimo FLOAT,
        timestamp INTEGER,
        fecha DATE)
''')
```

```python
# Realizamos una solicitud a la API de https://api.polygon.io para obtener los datos del activo "AAPL"
url = 'https://api.polygon.io/v2/aggs/ticker/AAPL/range/1/day/2022-01-01/2022-12-31'
params = {'apiKey': 'yDi5vVBSp38kzp_CwZle07pn_1komWBy'}
response = requests.get(url, params)
# Convertir la respuesta en un diccionario Python
data = json.loads(response.text)
print(response.text)
```

```python
# Convertimos el timestamp en un formato de fecha compatible con SQL con el fin de almacenarlo
from datetime import datetime
for d in data["results"]:
    timestamp = d["t"]
    date_object = datetime.fromtimestamp(timestamp/1000)
    fecha = date_object.strftime('%Y-%m-%d')
    cursor.execute('''
    INSERT INTO Stocks(stock, volumen, precio_promedio, precio_apertura, precio_cierre, precio_maximo, precio_minimo, timestamp, fecha)
    VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?)''',('APPL', d['v'], d['vw'], d['o'], d['c'], d['h'], d['l'], d['t'], fecha)
    )
# Comitimos los resultados a la base de datos
con.commit()
```
